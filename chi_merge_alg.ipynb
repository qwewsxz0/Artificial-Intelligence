# process2
## calculate chi square value (o-e)^2/e for every pair of rows
## where O is observed and E is expected values. 

```python
for i in range(0,10):
    e1 = df_processing.iloc[i]['Total']*df_processing.iloc[i]['obs_freq_of_c1_Adjacent_sum']/df_processing.iloc[i]['Total_Adjacent_sum']
    o1 = df_processing.iloc[i]['obs_freq_of_c1']
    val1 = (o1-e1)**2/e1
    print(val1)
    e2 = df_processing.iloc[i+1]['Total']*df_processing.iloc[i]['obs_freq_of_c1_Adjacent_sum']/df_processing.iloc[i]['Total_Adjacent_sum']
    o2 = df_processing.iloc[i+1]['obs_freq_of_c1']
    val2 = (o2-e2)**2/e2
    print(val2)
```

```python
chi_value_list = []
for i in range(0,10):
    chi_value = 0
    for name in ["c1","c2","c3"]:
        e1 = df_processing.iloc[i]['Total']*df_processing.iloc[i]['obs_freq_of_'+name+'_Adjacent_sum']/df_processing.iloc[i]['Total_Adjacent_sum']
        if e1<=0.5:
            e1=0.5
        else:
            e1 = e1
        o1 = df_processing.iloc[i]['obs_freq_of_'+name]
        val1 = (o1-e1)**2/e1
#         if i==0:
#             print('e1',e1)
#             print('o1',o1)
#             print('val1',val1)
        e2 = df_processing.iloc[i+1]['Total']*df_processing.iloc[i]['obs_freq_of_'+name+'_Adjacent_sum']/df_processing.iloc[i]['Total_Adjacent_sum']
        o2 = df_processing.iloc[i+1]['obs_freq_of_'+name]
        if e2<=0.5:
            e2=0.5
        else:
            e2 = e2
        val2 = (o2-e2)**2/e2
#         if i==0:
#             print('e2',e2)
#             print('o2',o2)
#             print('val2',val2)
            
        chi_value = chi_value+val1+val2
    chi_value_list.append(chi_value)
```

```python
finish_flag = 1
th = 4.61
value_of_A_num = [1.3,1.4,1.8,2.4,6.5,8.7,12.1,29.4,56.2,87.1,89.0]
value_of_A_str = []
for e in value_of_A_num:
    value_of_A_str.append(str(e))
c1 = [1,0,1,6,3,6,7,0,2,0,1]
c2 = [0,1,1,0,2,0,2,0,4,1,1]
c3 = [4,0,1,2,4,1,3,1,0,3,2]
dict_data = {"value_of_A":value_of_A_str,"obs_freq_of_c1":c1,"obs_freq_of_c2":c2,"obs_freq_of_c3":c3}
df = pd.DataFrame(dict_data)
df['Total'] = 0
df['Total'] = df['obs_freq_of_c1']+df['obs_freq_of_c2']+df['obs_freq_of_c3']

df_processing = df.copy()
df_ = df.copy()

while(finish_flag):
    df_processing = df.copy()
    row_number = len(df.index)
    ori_data_columns = df.columns[1:].tolist()
    process_name = []
    for name in ori_data_columns:
        df_processing[name+'_Adjacent_sum'] = 0
        process_name.append(name+'_Adjacent_sum')
        for i in range(0,row_number-1):
            df_processing[name+'_Adjacent_sum'][i] = df_processing[i:i+2][name].sum()
    chi_value_list = []
    for i in range(0,row_number-1):
        chi_value = 0
        for name in ["c1","c2","c3"]:
            e1 = df_processing.iloc[i]['Total']*df_processing.iloc[i]['obs_freq_of_'+name+'_Adjacent_sum']/df_processing.iloc[i]['Total_Adjacent_sum']
            if e1<=0.5:
                e1=0.5
            else:
                e1 = e1
            o1 = df_processing.iloc[i]['obs_freq_of_'+name]
            val1 = (o1-e1)**2/e1
    #         if i==0:
    #             print('e1',e1)
    #             print('o1',o1)
    #             print('val1',val1)
            e2 = df_processing.iloc[i+1]['Total']*df_processing.iloc[i]['obs_freq_of_'+name+'_Adjacent_sum']/df_processing.iloc[i]['Total_Adjacent_sum']
            o2 = df_processing.iloc[i+1]['obs_freq_of_'+name]
            if e2<=0.5:
                e2=0.5
            else:
                e2 = e2
            val2 = (o2-e2)**2/e2
    #         if i==0:
    #             print('e2',e2)
    #             print('o2',o2)
    #             print('val2',val2)

            chi_value = chi_value+val1+val2
        chi_value_list.append(chi_value)

    if min(chi_value_list)<th:
        print('need merge')
        print(chi_value_list.index(min(chi_value_list)))
        merge_interval_index = chi_value_list.index(min(chi_value_list))
        newrow = df.iloc[merge_interval_index]+df.iloc[merge_interval_index+1]
        df.iloc[merge_interval_index] = newrow
        df = df.drop(index=merge_interval_index+1).reset_index(drop=True)

    else:
        print('finish')
        finish_flag = 0
```
